# Scraping

```
base = 'https://juba.scba.gov.ar/VerTextoCompleto.aspx?idFallo='
urls = []
#textos =  list(range(159930, 161930)) # primera extracción
textos =  list(range(161930, 164229)) # segunda extracción

for texto in textos:
    urls.append(base + str(texto))

print(urls)
```

```
len(urls)
```

```
def url_get_text(url):
  '''Obtener los textos '''    
  html = requests.get(url)
  soup = bs4.BeautifulSoup(html.text, "html.parser")
  
  for script in soup(["script", "style"]):
    script.extract()    # rip it out
  
  text = soup.get_text()

  # break into lines and remove leading and trailing space on each
  lines = (line.strip() for line in text.splitlines())
  # break multi-headlines into a line each
  chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
  # drop blank lines
  result = '\n'.join(chunk for chunk in chunks if chunk)
  
  return result
 
```

```
# Recorrer las URLs y obtener los textos
resultado = []

for i in range(len(urls)):
  u = urls[i]  
  resultado.append(url_get_text(u))

print(len(resultado))
```
